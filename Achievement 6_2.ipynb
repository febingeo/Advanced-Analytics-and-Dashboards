#### Scatterplots

```python
# Create a scatterplot for the "house age" and "price per unit" columns in seaborn

sns.lmplot(x = 'Year Built', y = 'Number of views last 7 days', data = sub)
```

```python
fig = plt.figure(figsize=(20,15) )
ax1 = fig.add_subplot(223)
sns.regplot(x='Year Built', y='Number of views last 7 days', data=sub, line_kws={'color': 'red'}, ci=None, x_jitter=0.1)
ax1.set_title("Year Built and  Number of views last 7 days ")
ax2 = fig.add_subplot(224)
sns.distplot(sub['Year Built'])
ax2.set_title("Year Built Distribution")
```

#### In the scatterplot, you can see that while there is, indeed, a weak negative relationship, the trend line isn’t very steep, and there are many points far away from this line

#### Relationship betweeen YEAR BUILT and  Number of views last 7 days are non - linear

```python
fig = plt.figure(figsize=(20,15) )
ax1 = fig.add_subplot(223)
sns.regplot(x='Length', y='Width', data=sub, line_kws={'color': 'red'}, ci=None, x_jitter=0.1)
ax1.set_title("Length and Width ")
#ax2 = fig.add_subplot(224)
#sns.distplot(sub['Year Built'])
#ax2.set_title("Year Built Distribution")
```

### Categorical Plots

```python
# Use a histogram to visualize the distribution of the variable NUMBER OF VIEWS LAST 7 DAYS
sns.histplot(sub['Number of views last 7 days'], bins = 20, kde = True)
```

```python
sub.loc[sub['Number of views last 7 days'] < 100, 'Views'] = 'Less Views'
sub.loc[(sub['Number of views last 7 days'] >= 100) & (sub['Number of views last 7 days'] < 300), 'Views'] = 'Moderate views'
sub.loc[sub['Number of views last 7 days'] > 300, 'Views'] = 'High Views'
```
